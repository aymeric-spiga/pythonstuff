### Généralités

Nous définissons deux fonctions qui résolvent soit une (`euler1`), soit plusieurs (`euler2`), équation(s) différentielle(s) du premier ordre par la méthode d'Euler étant donnés :

  * une fonction du temps $t$ notée par exemple $u$ et que l'on cherche à déterminer d'après l'équation différentielle
  * une fonction $F$ telle que $u' = \frac{\textrm{d}u}{\textrm{d}t} = F(u(t)) \,$ définissant l'équation différentielle
  * une condition initiale $U_0 = u(t_i)$
  * une discrétisation de l'axe des temps donné par un temps initial $t_i$, un temps final $t_f$ et un nombre d'itérations temporelles $N$ entre $t_i$ et $t_f$ ; le pas de temps $\Delta t$ est donc $(t_f-t_i)/N$.

Dans le cas de `euler1` ci-dessous, $F$ est une fonction de réels $\mathbb{R}\rightarrow\mathbb{R}$ ($u$ et $U_0$ sont des réels) ; dans le cas `euler2`, u est une fonction de vecteurs $\mathbb{R}^{n}\rightarrow\mathbb{R}^{n}$ ($u$ et $U_0$ sont des vecteurs de réels).

La méthode d'Euler peut se formuler d'après la décomposition de Taylor au premier ordre comme la détermination des N termes de la suite $(u_k)_{k=0 \ldots N}$ telle que $$ u_{k+1} =  u_k + F(u_k) \, \Delta t \qquad \textrm{avec} \qquad u_0 = U_0 = u(t_i) \quad \textrm{et} \quad u_N = u(t_f)$$


```python
def euler1(F,U0,ti,tf,N):
  ## cas d'une seule équation du premier ordre
  ## --> U0 et l'argument de F sont des réels
  dt = (tf-ti)/N
  print "pas",dt
  t = np.linspace(ti,tf,N+1)
  u = np.empty(N+1)
  u[0] = U0
  for k in range(N):
    u[k+1] = u[k] + F(u[k])*dt
  return t,u
```

```python
def euler2(F,U0,ti,tf,N):
  ## cas de plusieurs équations du premier ordre
  ## --> U0 et l'argument de F sont des arrays de rang 1 (vecteurs)  
  m = np.size(U0)
  dt = (tf-ti)/N
  print "pas",dt
  t = np.linspace(ti,tf,N+1)
  u = np.empty([N+1,m])
  u[0,:] = U0
  for k in range(N):
    du = F(u[k,:])
    u[k+1,:] = u[k,:] + du*dt
  return t,u
```

**<font color='red'>Dans ce qui suit, nous utilisons la méthode 1 lorsque le problème peut se ramener à une seule équation différentielle du premier ordre. Dans ce cas, la méthode 2 fonctionne également comme illustré dans l'exemple simple *Décroissance radioactive*. En revanche, lorsque le problème fait intervenir plusieurs équations différentelles du second ordre, seule la méthode 2 peut fonctionner.</font>**

### Décroissance radioactive

Equation $N' = - \frac{N}{\tau}$

$N(t)$ concentration en espèce, $\tau$ temps de demi-vie

On pose $ u = N \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = N' = - \frac{N}{\tau} = F(u) \quad\textrm{avec}\quad F(u) = -\frac{u}{\tau} $

#### Méthode 1 : avec `euler1` 
Cette méthode travaille avec des réels. Ici, dans la mesure où nous avons une seule équation du premier ordre, elle est toute indiquée.

```python
## constante
tau = 0.2
U0 = 1000.
ti,tf,N = 0.,1.,5000
```

```python
def F(u):
  du = -u/tau
  return du
```

<small> NB: On peut définir également la fonction `F` en une ligne de python `F = lambda x:-x/tau` </small>

```python
## appel et plot
t,u = euler1(F,U0,ti,tf,N) ; plt.plot(t,u)
```

#### Méthode 2 : avec `euler2`
Cette méthode travaille avec des vecteurs. Ici, évidemment, elle est un peu artificielle (utilisation de vecteurs à un élément), mais nous l'illustrons à des fins d'introduction et de comparaison avec la méthode 1 dans un cas simple.

```python
## constante
tau = 0.2
U0 = [1000.]
ti,tf,N = 0.,1.,5000
```

```python
def F(u):
  du = np.empty(1)
  du[0] = -u[0]/tau
  return du
```

```python
## appel et plot
t,u = euler2(F,U0,ti,tf,N) ; plt.plot(t,u[:,0])
```

### Cinétique chimique d'ordre 2

Equation $c' = k \, (1-c)^2$

$c(t)$ concentration en espèce, $k$ constante de réaction

On pose $ u = c \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = c' = k \, (1-c)^2 = F(u) \quad\textrm{avec}\quad F(u) = k \, (1-u)^2 $

```python
## constante
k = 1e-2
U0 = 1e-3
ti,tf,N = 0.,500.,5000
```

```python
def F(u):
  return k*(1-u)**2
```

```python
## appel et plot
t,u = euler1(F,U0,ti,tf,N) ; plt.plot(t,u)
```

### Modèle logistique (limitation de croissance d'une population par milieu)

Equation $P' = \alpha \, P \, (1-\frac{P}{P_l})$

$P(t)$ population, $\alpha$ vitesse de croissance, $P_l$ population limite

On pose $ u = P \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = P' = \alpha \, P \, (1-\frac{P}{P_l}) = F(u) \quad\textrm{avec}\quad F(u) = \alpha \, u \, (1-\frac{u}{P_l})$

```python
## constante
alpha = 0.5
Pl = 12000.
U0 = 1e-3
ti,tf,N = 0.,50.,5000
```

```python
def F(u):
  return alpha*u*(1-(u/Pl))
```

```python
## appel et plot
t,u = euler1(F,U0,ti,tf,N) ; plt.plot(t,u)
```

### Equation de Torricelli

Equation $A(h) \, h' = - \beta \sqrt{h}$

$h(t)$ hauteur du fluide dans récipient, $\beta$ facteur décrivant le trou à la base du récipient, $A(h)$ facteur de forme du récipient (variation de l'aire de la section du récipient en fonction de la hauteur)

On pose $ u = h \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = h' = - \frac{k \, \sqrt{h}}{A(h)} = F(u) \quad\textrm{avec}\quad F(u) = - \frac{k \, \sqrt{u}}{A(u)} $

```python
## constante
beta = 1.
U0 = 20.e-2
ti,tf,N = 0.,0.99,5000
# fonction A(u) 
# -- variation de l'aire en fonction hauteur liquide
# -- supposons un récipient très évasé (exponentielle)
def A(h):
  return np.exp(h)
```

```python
def F(u):
  return -beta*np.sqrt(u)/A(u)
```

```python
## appel et plot
t,u = euler1(F,U0,ti,tf,N) ; plt.plot(t,u)
```

### Chute d'une bille dans l'huile

Equation $z'' - \kappa \, z' + m \, g = 0 \qquad \kappa = 6 \, \pi \, R \, \eta$

$z(t)$ altitude de la bille, $m$ masse de la bille, $g$ accélération de la pesanteur, $R$ rayon de la bille, $\eta$ viscosité dynamique de l'huile

On pose $ u = z' \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = z'' = \kappa \, z' - m \, g = F(u) \quad\textrm{avec}\quad F(u) = \kappa \, u - m\, g $

```python
## constantes
R,m,g = 3e-2,1e-3,9.81
eta = 0.1 # huile olive
kappa = 6.*np.pi*R*eta
U0 = 0.
ti,tf,N = 0.,100.,5000
```

```python
def F(u):
  return (kappa*u)-(m*g)
```

```python
## appel et plot
t,u = euler1(F,U0,ti,tf,N) ; plt.plot(t,u)
```

### Oscillateur harmonique

Equation $\theta'' + k^2 \, \sin \theta = 0 \qquad k = \sqrt{\frac{g}{\ell}} \quad \textrm{ou} \quad k = \sqrt{\frac{m\,g}{J}}$

$\theta(t)$ angle entre pendule et verticale, $g$ accélération de la pesanteur, $\ell$ longueur de la ficelle

$\theta(t)$ angle entre pendule et verticale, $g$ accélération de la pesanteur, $m$ masse du pendule pesant, $J$ moment d'inertie

Vecteur $ u = \begin{bmatrix} \theta \\ \theta' \end{bmatrix} \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = \begin{bmatrix} \theta' \\ \theta'' \end{bmatrix} = \begin{bmatrix} \theta' \\ -k^2 \, \sin \theta \end{bmatrix} = F(u) \quad\textrm{avec}\quad F(u) = \begin{bmatrix} u[1] \\ -k^2 \, \sin u[0] \end{bmatrix} $

```python
## constantes
g,m,J = 10.,10.,100.
k = np.sqrt(m*g/J)
theta0,omega0 = np.pi/4.,0.
U0 = [theta0,omega0]
Tp = 2*np.pi/k
ti,tf,N = 0.,3.*Tp,5000
```

```python
def F(u):
  du = np.empty(2)
  du[0] = u[1]
  du[1] = -k*k*np.sin(u[0])
  return du
```

```python
## appel et plot
t,u = euler2(F,U0,ti,tf,N) ; plt.plot(t,u[:,0])
```

### Oscillateur harmonique avec amortissement

Equation $x'' + \frac{2}{\tau} \, x' + {\omega_0}^2 \, x = 0$

$x(t)$ signal mesuré, $\tau$ temps d'amortissement, $\omega_0$ fréquence d'oscillation

Vecteur $ u = \begin{bmatrix} x \\ x' \end{bmatrix} \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = \begin{bmatrix} x' \\ x'' \end{bmatrix} = \begin{bmatrix} x' \\ -\frac{2}{\tau} \,x'-{\omega_0}^2 \, x \end{bmatrix} = F(u) \quad\textrm{avec}\quad F(u) = \begin{bmatrix} u[1] \\ -\frac{2}{\tau} \, u[1] - {\omega_0}^2 \, u[0] \end{bmatrix} $

```python
## constantes
U0 = [-1.,0.]
tau,omega = 10.,1. # omega > 1/tau -- sous-critique
ti,tf,N = 0.,5.*tau,5000
#tau,omega = 0.1,1. # omega < 1/tau -- sur-critique
#ti,tf,N = 0.,1000.*tau,5000
```

```python
def F(u):
  du = np.empty(2)
  du[0] = u[1]
  du[1] = -(2.*u[1]/tau)-(omega*omega*u[0])
  return du
```

```python
## appel et plot
t,u = euler2(F,U0,ti,tf,N) ; plt.plot(t,u[:,0])
```

### Elargissement d'un faisceau électronique (sans champ extérieur)

Equation $y'' = \frac{A}{y}$

$y(t)$ position du faisceau, $A$ facteur d'élargissement

Vecteur $ u = \begin{bmatrix} y \\ y' \end{bmatrix} \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = \begin{bmatrix} y' \\ y'' \end{bmatrix} = \begin{bmatrix} y' \\ -A/y \end{bmatrix} = F(u) \quad\textrm{avec}\quad F(u) = \begin{bmatrix} u[1] \\ -A/u[0] \end{bmatrix} $

```python
## constantes
A = 100.
y0,yprime0 = 10.,10
U0 = [y0,yprime0]
ti,tf,N = 0.,3.*Tp,5000
```

```python
def F(u):
  du = np.empty(2)
  du[0] = u[1]
  du[1] = -A/u[0]
  return du
```

```python
## appel et plot
t,u = euler2(F,U0,ti,tf,N) ; plt.plot(t,u[:,0])
```

### Modèle prédateur-proie

Equation $x' = x \, (1-y) \qquad y' = \mu \, y \,(x-1)$

$y(t)$ population des prédateurs, $x(t)$ population des proies, $\mu$ facteur d'interaction

Vecteur $ u = \begin{bmatrix} x \\ y \end{bmatrix} \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = \begin{bmatrix} x' \\ y' \end{bmatrix} = \begin{bmatrix} x \, (1-y) \\ \mu \, y \,(x-1) \end{bmatrix} = F(u) \quad\textrm{avec}\quad F(u) = \begin{bmatrix} u[0] \, (1-u[1]) \\ \mu \, u[1] \, (u[0]-1) \end{bmatrix} $

```python
## constantes
mu = 2
x0,y0 = 20,20
U0 = [x0,y0]
ti,tf,N = 0.,1.,1000
```

```python
def F(u):
  du = np.empty(2)
  du[0] = u[0]*(1-u[1])
  du[1] = mu*u[1]*(u[0]-1) 
  return du
```

```python
## appel et plot
t,u = euler2(F,U0,ti,tf,N) ; plt.plot(t,u[:,0],'g.',t,u[:,1],'r.')
```

### Modèle de Sel-kov d'oscillation de la glycolyse

Equation $$x' = -x + a\, y + x^2\,y \qquad y' = b - a \, y + x^2 \, y$$

$y(t)$ concentration F6P, $x(t)$ concentration ADP, $a$ et $b$ constantes de couplage positives 

Vecteur $ u = \begin{bmatrix} x \\ y \end{bmatrix} \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = \begin{bmatrix} x' \\ y' \end{bmatrix} = \begin{bmatrix} -x + a\, y + x^2\,y \\ b - a \, y + x^2 \, y \end{bmatrix} = F(u) \quad\textrm{avec}\quad F(u) = \begin{bmatrix} -u[0] + a\, u[1] + u[0]^2\,u[1] \\ b - a \, u[1] + u[0]^2 \, u[1] \end{bmatrix} $

```python
## constantes
a,b = 1e-6,1e-3
x0,y0 = 1e-4,1e-6
U0 = [x0,y0]
ti,tf,N = 0.,1.,10000
```

```python
def F(u):
  du = np.empty(2)
  du[0] = -u[0]+a*u[1]+u[0]*u[0]*u[1]
  du[1] = b-a*u[1]+u[0]*u[0]*u[1]
  return du
```

```python
## appel et plot
t,u = euler2(F,U0,ti,tf,N) ; plt.plot(t,u[:,0],'b.',t,u[:,1],'r.')
```

### Modèle de chaos déterministe de Lorenz

Equation $X' = - \sigma \, X + \sigma \, Y \qquad Y' = - X \, Z + r \, X - Y \qquad Z' = X \, Y - b \, Z \qquad$ 

$X(t),Y(t),Z(t)$ représentent l'état de l'atmosphère et $\sigma$,$r$,$b$ constantes. Lorenz dans son article de 1963 se place dans le cas $\sigma=10$, $b=8/3$, $r=28$. 

Vecteur $ u = \begin{bmatrix} X \\ Y \\ Z \end{bmatrix} \qquad$ Trouver $F$ tel que $u_{k+1} = u_k + F(u_{k}) \, \Delta t$

Vecteur dérivé $ u' = \begin{bmatrix} X' \\ Y' \\ Z' \end{bmatrix} = \begin{bmatrix} - \sigma \, X + \sigma \, Y \\ - X \, Z + r \, X - Y \\ X \, Y - b \, Z \end{bmatrix} = F(u) \quad\textrm{avec}\quad F(u) = \begin{bmatrix} - \sigma \, u[0] + \sigma \, u[1] \\ - u[0] \, u[2] + r \, u[0] - u[1] \\ u[0] \, u[1] - b \, u[2] \end{bmatrix} $

```python
## constantes
sigma,b,r = 10,8./3.,28.
x0,y0,z0 = -15,0,15
U0 = [x0,y0,z0]
ti,tf,N = 0.,20.,1000
```

```python
def F(u):
  du = np.empty(3)
  du[0] = -sigma*u[0]+sigma*u[1]
  du[1] = -u[0]*u[2]+r*u[0]-u[1]
  du[2] = u[0]*u[1]-b*u[2]
  return du
```

```python
## appel et plot
t,u = euler2(F,U0,ti,tf,N) ; plt.plot(t,u[:,0],'g.',t,u[:,1],'r.',t,u[:,2],'b.')
```

```python
# version 3D de la figure
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(u[:,0],u[:,1],u[:,2])
plt.show()
```

### Utilisation de `odeint`

Avant de commencer à disserter sur l'utilisation de `odeint`, revenons sur `euler2` que nous pouvons modifier en `euler2gen` pour tendre vers une plus grande généralité en

* acceptant virtuellement n'importe laquelle des discrétisations en temps (y compris non régulières)
* adoptant la notation de Cauchy générale d'une équation différentielle du premier ordre
$u' = \frac{\textrm{d}u}{\textrm{d}t} = F(u(t),t) \,$

Comparer ainsi

```python
def euler2(F,U0,ti,tf,N):
  ## cas de plusieurs équations du premier ordre
  ## --> U0 et l'argument de F sont des arrays de rang 1 (vecteurs)  
  m = np.size(U0)
  dt = (tf-ti)/N
  print "pas",dt
  t = np.linspace(ti,tf,N+1)
  u = np.empty([N+1,m])
  u[0,:] = U0
  for k in range(N):
    du = F(u[k,:])
    u[k+1,:] = u[k,:] + du*dt
  return t,u
```

à

```python
def euler2gen(F,U0,t):
  ## cas de plusieurs équations du premier ordre
  ## --> U0 et l'argument de F sont des arrays de rang 1 (vecteurs)   
  ## t doit être un tableau de N éléments contenant les itérations en temps
  ## --> F prend comme argument supplémentaire un réel, l'instant t[k] (cf. notation de Cauchy)
  m = np.size(U0)
  u = np.empty([N+1,m])
  u[0,:] = U0
  for k in range(N):
    du = F(u[k,:],t[k])
    dt = t[k+1]-t[k]
    u[k+1,:] = u[k,:] + du*dt
  return u # noter qu'il n'est plus nécessaire de sortir t qui est défini hors de euler2gen
```

Résolvons de nouveau le problème du pendule avec `euler2gen`

```python
## constantes
g,m,J = 10.,10.,100.
k = np.sqrt(m*g/J)
theta0,omega0 = np.pi/4.,0.
U0 = [theta0,omega0]
Tp = 2*np.pi/k
ti,tf,N = 0.,3.*Tp,5000
```

```python
def F(u,t):
  du = np.empty(2)
  du[0] = u[1]
  du[1] = -k*k*np.sin(u[0])
  return du
```

Noter que `t` est une variable muette de `F` pour respecter la convention générale de `euler2gen`. Cela paraît artificiel ici, et parfaitement inutile, mais cette convention générale est adoptée par `odeint` donc nous nous y préparons.

```python
t = np.linspace(ti,tf,N+1)
u = euler2gen(F,U0,t) ; plt.plot(t,u[:,0])
```

Nous sommes donc prêts à utiliser `odeint`. Cette fonction du package `scipy.integrate` réalise exactement la même chose que `euler2gen` à ceci près qu'elle emploie des discrétisations plus robustes (comprendre, stables), plus précises et optimisées que la méthode d'Euler.

Son usage est tout à fait similaire à `euler2gen`

```python
from scipy.integrate import odeint
t = np.linspace(ti,tf,N+1)
u = odeint(F,U0,t) ; plt.plot(t,u[:,0])
```

Comparons les 2 méthodes pour un nombre d'itérations $N$ faible (20). Nous allons illustrer ainsi la robustesse de `odeint` par rapport à une méthode d'Euler simple.

```python
N = 30
t = np.linspace(ti,tf,N+1)
u_euler = euler2gen(F,U0,t)
u_odeint = odeint(F,U0,t)
plt.plot(t,u_euler[:,0],'g-',label="Euler (aaargh)")
plt.plot(t,u_odeint[:,0],'b-',label="odeint (dans les clous)")
plt.legend(loc='upper left')
```

### Exemple complet d'utilisation de `odeint`

Récapitulons l'utilisation de `odeint` pour l'exemple *Chute d'une bille dans l'huile*.

1) Définition des constantes

```python
R,m,g = 3e-2,1e-3,9.81
eta = 0.1 # huile olive
kappa = 6.*np.pi*R*eta
U0 = 0.
ti,tf,N = 0.,100.,5000
```

2) Définition de la fonction $F$
Ne pas oublier la variable t, qu'elle soit effectivement utilisée ou non

```python
def F(u,t):
  return (kappa*u)-(m*g)
```

3) Définition de la grille temporelle

```python
t = np.linspace(ti,tf,N+1)
```

4) Appel de odeint

```python
u = odeint(F,U0,t)
```

5) Création d'une figure

```python
plt.plot(t,u[:,0])
plt.ylabel("Vitesse de chute dans l'huile (m/s)")
plt.xlabel("Temps (s)")
plt.title("Chute d'une goutte dans l'huile")
```
